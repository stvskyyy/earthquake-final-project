# Data Inspection

This notebook’s goal is only to answer:

- What sheets are inside KODA_Datasets.xlsx?
- What columns exist?
- How many rows are there?
- What are the missing values?
- Which columns are useful for modeling?

```python
import pandas as pd
import os
from pathlib import Path

RAW_DATA_DIR = Path("../data/raw")

csv_files = list(RAW_DATA_DIR.glob("*.csv"))

print(f"Found {len(csv_files)} CSV files:")
for file in csv_files:
    print(file.name)
```

```python
datasets = {}

for file in csv_files:
    try:
        df = pd.read_csv(file)
        datasets[file.stem] = df
        
        print(f"Loaded: {file.name}")
        print(f"Shape: {df.shape}")
        print("-" * 60)
        
    except Exception as e:
        print(f"Error loading {file.name}: {e}")
```

```python
for name, df in datasets.items():
    print(f"\n===== {name} =====")
    display(df.head())
```

```python
for name, df in datasets.items():
    print(f"\n===== {name} =====")
    
    print("Shape:")
    print(df.shape)
    
    print("\nColumns:")
    print(df.columns.tolist())
    
    print("\nData Types:")
    print(df.dtypes)
    
    print("\nMissing Values:")
    print(df.isna().sum())
    
    print("\nDuplicate Rows:")
    print(df.duplicated().sum())
    
    print("-" * 80)
```

```python
inventory = []

for name, df in datasets.items():
    inventory.append({
        "dataset_name": name,
        "rows": df.shape[0],
        "columns": df.shape[1],
        "column_names": ", ".join(df.columns.astype(str)),
        "missing_values_total": df.isna().sum().sum(),
        "duplicate_rows": df.duplicated().sum()
    })

inventory_df = pd.DataFrame(inventory)
inventory_df
```

```python
OUTPUT_DIR = Path("../outputs/tables")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

inventory_df.to_csv(OUTPUT_DIR / "dataset_inventory.csv", index=False)

print("Saved dataset inventory to outputs/tables/dataset_inventory.csv")
```
